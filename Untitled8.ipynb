```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and prepare data
#file_path = '/mnt/data/sample_email_data.csv'
email_data = pd.read_csv('sample_email_data.csv')
email_data['Timestamp'] = pd.to_datetime(email_data['Timestamp'])
```

```python
# Filter phishing emails and extract sender domains
phishing_emails = email_data[email_data['Category'] == 'Phishing']
phishing_emails['Sender_Domain'] = phishing_emails['Sender_Email'].apply(lambda x: x.split('@')[1])

# Count occurrences of each sender domain in phishing emails
phishing_domains = phishing_emails['Sender_Domain'].value_counts().head(10)
print("Top Phishing Sender Domains:\n", phishing_domains)

# Plot
phishing_domains.plot(kind='bar', title="Top Phishing Sender Domains", ylabel="Count", xlabel="Sender Domain")
plt.show()
```

```python
# Filter for emails with large attachments in phishing/spam categories
suspicious_attachments = email_data[(email_data['Attachment_Size'] > 5) &
                                    (email_data['Category'].isin(['Phishing', 'Spam']))]
print("Suspicious Attachments in Phishing/Spam:\n", suspicious_attachments[['Sender_Email', 'Attachment_Size']])

# Plot distribution of attachment sizes
suspicious_attachments['Attachment_Size'].plot(kind='hist', bins=10, title="Attachment Size Distribution (MB)")
plt.xlabel("Attachment Size (MB)")
plt.show()
```

```python
# Group by month and count email activity
monthly_email_counts = email_data.resample('ME', on='Timestamp').size()
print("Monthly Email Volume:\n", monthly_email_counts)

# Plot monthly email trends
monthly_email_counts.plot(kind='line', title="Monthly Email Volume", ylabel="Number of Emails", xlabel="Month")
plt.show()
```

```python
# Identify internal and external emails based on company domain (example: "company.com")
internal_domain = "company.com"
internal_emails = email_data[email_data['Sender_Email'].str.endswith(internal_domain) &
                             email_data['Recipient_Email'].str.endswith(internal_domain)]
external_emails = email_data[~email_data.index.isin(internal_emails.index)]

# Count and print internal vs. external emails
print("Internal Emails:", len(internal_emails))
print("External Emails:", len(external_emails))
```

```python
# Filter flagged, blocked, and quarantined emails
security_risk_emails = email_data[email_data['Status'].isin(['Flagged', 'Blocked', 'Quarantined'])]

# Group and plot security risks by month
security_risk_trends = security_risk_emails.resample('ME', on='Timestamp').size()
security_risk_trends.plot(kind='line', title="Monthly Security Risk Emails", ylabel="Count", xlabel="Month")
plt.show()
```

```python
# Count the categories within security-risk emails
security_category_counts = security_risk_emails['Category'].value_counts()
print("Security Risks by Category:\n", security_category_counts)

# Plot
security_category_counts.plot(kind='pie', autopct='%1.1f%%', title="Security Risk Categories")
plt.ylabel("")
plt.show()
```

```python
# Sent vs. Received
status_counts = email_data['Status'].value_counts()
print("Email Status Counts:\n", status_counts)

# Plot
status_counts.plot(kind='bar', title="Sent vs. Received Email Counts", ylabel="Count", xlabel="Status")
plt.show()
```

```python
# Distribution of Attachment Sizes across all emails
email_data['Attachment_Size'].plot(kind='hist', bins=15, title="Attachment Size Distribution", xlabel="Attachment Size (MB)")
plt.show()
```
